# INIT METHOD
All classes have a built-in method called __init__(), which is always executed when the class is being initiated.

The __init__() method is used to assign values to object properties, or to perform operations that are necessary when the object is being created.

```python
# class Person:
#   pass

# p1 = Person()
# p1.name = "Tobias"
# p1.age = 25

# print(p1.name)
# print(p1.age)

# print(Person.name)
# This will show error because person has no attribute name its defined
# for the object only
```

```python
class Person:
    def __init__(self,name,age):
        #     def __init__(self,name,age = 19 this sets the default value):
        self.name = name
        self.age = age
        # print(self.name)
    # self is a defult argument in the init method
    # it makes setting property for class easy
p1 = Person("Veer",18)
print(p1.name)
```

# SELF PARAMETER
Self is a reference to the current instance of the class
It is used to acess the properties and methods that belong to the class

It must be the first parameter of any of the methods of the class

The self parameter links the method to the specific object and need not be named self

```python
class Person:
  def __init__(obj, name):
    obj.name = name

  def printname(raju):
    print(raju.name)
  def selfcalling(rave):
    rave.printname()
    # You can call one method in other method using self
p1 = Person("Veer")
p2 = Person("Ravi")
# print(Person)
p1.selfcalling()
p2.selfcalling()
```
